```python
sports_v2['Q22'] = sports_v2['Q22'].astype('int')
```

```python
conditions = [
    (sports_v2['Q22'] <= 25),
    (sports_v2['Q22'] >= 26) & (sports_v2['Q22'] <= 40),
    (sports_v2['Q22'] >= 41) & (sports_v2['Q22'] <= 55),
    (sports_v2['Q22'] >= 56) & (sports_v2['Q22'] <= 74),
    (sports_v2['Q22'] >= 75)
    ]

values = ['Gen Z', 'Millennial', 'Gen X', 'Boomer', 'Silent Gen']

sports_v2['generation'] = np.select(conditions, values)
```

```python
sports_v2.iloc[:,0:3] = sports_v2.iloc[:,0:3].astype('int')
sports_v2.iloc[:,4:-1] = sports_v2.iloc[:,4:-1].astype('int')
sports_v2.dtypes
```

```python
# Define 'gender'
conditions1 = [
    (sports_v2['Q23'] == 1),
    (sports_v2['Q23'] == 2), 
    (sports_v2['Q22'] == 3),
    (sports_v2['Q23'] == 4)
    ]

values1 = ['Male', 'Female', 'Non-binary', 'Null']

sports_v2['gender'] = np.select(conditions1, values1)
```

```python
# Define 'race'
conditions2 = [
    (sports_v2['Q24'] == 1),
    (sports_v2['Q24'] == 2), 
    (sports_v2['Q24'] == 3),
    (sports_v2['Q24'] == 4),
    (sports_v2['Q24'] == 5),
    (sports_v2['Q24'] == 6)
    ]

values2 = ['White', 'African American', 'Hispanic', 'Asian', 'Mixed Race', 'Other']

sports_v2['race'] = np.select(conditions2, values2)
```

```python
# Define 'SES'
conditions3 = [
    (sports_v2['Q25'] == 1),
    (sports_v2['Q25'] == 2), 
    (sports_v2['Q25'] == 3),
    (sports_v2['Q25'] == 4)
    ]

values3 = ['<50k', '>50k & <100k', '>100k & <200k', '>200k']

sports_v2['SES'] = np.select(conditions3, values3)
```

```python
# Define 'political_spectrum'
conditions4 = [
    (sports_v2['Q26_1'] == 1),
    (sports_v2['Q26_1'] == 2), 
    (sports_v2['Q26_1'] == 3),
    (sports_v2['Q26_1'] == 4),
    (sports_v2['Q26_1'] == 5),
    (sports_v2['Q26_1'] == 6),
    (sports_v2['Q26_1'] == 7)
    ]

values4 = ['Extremely Conservative', 'Moderately Conservative', 'Quite Conservative', 'Neutral', 'Quite Liberal', 'Moderately Conservative', 'Extremely Liberal']

sports_v2['political_spectrum'] = np.select(conditions4, values4)
```

```python
# Define 'religious spectrum'
conditions5 = [
    (sports_v2['Q26_2'] == 1),
    (sports_v2['Q26_2'] == 2), 
    (sports_v2['Q26_2'] == 3),
    (sports_v2['Q26_2'] == 4),
    (sports_v2['Q26_2'] == 5),
    (sports_v2['Q26_2'] == 6),
    (sports_v2['Q26_2'] == 7)
    ]

values5 = ['Extremely Religious', 'Moderately Religious', 'Quite Religious', 'Neutral', 'Quite Non-religious', 'Usually Non-religious', 'Extremely Non-religious']

sports_v2['religious_spectrum'] = np.select(conditions5, values5)
```

```python
# Define 'nationality'
conditions6 = [
    (sports_v2['Q27'] == 1),
    (sports_v2['Q27'] == 2), 
    (sports_v2['Q27'] == 3),
    (sports_v2['Q27'] == 4),
    (sports_v2['Q27'] == 5),
    (sports_v2['Q27'] == 6),
    (sports_v2['Q27'] == 7)
    ]

values6 = ['United States', 'Canada', 'Mexico', 'Latin America', 'Europe', 'Africa', 'Asia/Oceania']

sports_v2['nationality'] = np.select(conditions6, values6)
```
